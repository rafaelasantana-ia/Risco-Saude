# 📊 Análise Exploratória de Dados (EDA) - Risco Saúde
Este notebook tem como foco entender profundamente o comportamento das variáveis do dataset `dataset_saude_brasil.csv`.
Aqui faremos:
1. Verificação de dados faltantes (Nulos).
2. Entendimento das distribuições estatísticas.
3. Análise bivariada e correlação entre os hábitos e o Risco de Doença.

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configuração de Estilo Visual
sns.set_theme(style="whitegrid", palette="muted")

# Carregar Dados
df = pd.read_csv('dataset_saude_brasil.csv')
print(f'Linhas: {df.shape[0]} | Colunas: {df.shape[1]}')
df.head()
```

## 1. Estatística Descritiva e Nulos

```
# Visualizando integridade dos dados
display(df.info())
print("\n-- Valores Nulos --")
display(df.isnull().sum())
print("\n-- Estatística das Variáveis Numéricas --")
display(df.describe())
```

## 2. Distribuição das Variáveis Alvo e Clínicas

```
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

sns.countplot(data=df, x='Risco_Doenca', order=['Baixo', 'Moderado', 'Alto', 'Muito Alto'], palette='viridis', ax=axes[0])
axes[0].set_title('Distribuição de Risco de Doença')

sns.histplot(df['Idade'], bins=30, kde=True, color='skyblue', ax=axes[1])
axes[1].set_title('Distribuição de Idade')

sns.histplot(df['IMC'], bins=30, kde=True, color='salmon', ax=axes[2])
axes[2].set_title('Distribuição de IMC')

plt.tight_layout()
plt.show()
```

## 3. Impacto de Hábitos no Risco Cardíaco
Vamos analisar como o fato de Fumar ou beber afeta a proporção de risco.

```
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Tabagismo
sns.countplot(data=df, x='Risco_Doenca', hue='Fumante', palette='Set2', 
              order=['Baixo', 'Moderado', 'Alto', 'Muito Alto'], ax=axes[0])
axes[0].set_title('Risco de Doença por Tabagismo')

# Álcool
sns.countplot(data=df, x='Risco_Doenca', hue='Alcool', palette='Set1', 
              order=['Baixo', 'Moderado', 'Alto', 'Muito Alto'], ax=axes[1])
axes[1].set_title('Risco de Doença por Consumo de Álcool')

plt.tight_layout()
plt.show()
```

## 4. Mapa de Correlações Numéricas
Ajuda a ver como colesterol, pressão e outros se influenciam matematicamente.

```
plt.figure(figsize=(10, 8))
# Seleciona apenas colunas numéricas
df_num = df.select_dtypes(include=[np.number])

# Para o Heatmap, vamos temporariamente converter algumas nulas
df_num = df_num.fillna(df_num.median())

corr = df_num.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Heatmap de Correlação')
plt.show()
```
